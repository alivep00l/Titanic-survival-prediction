```python
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.info()
```

```python
newdf = df.dropna(subset = ['Age'])
```

```python
newdf.info()
```

```python
newdf.head()
```

```python
input = newdf.drop(columns=['Survived','PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'])
output = newdf['Survived']
```

```python
input.head()
```

```python
output.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
input['Sex_n']= le_sex.fit_transform(input['Sex'])
```

```python
input.drop('Sex', inplace = True, axis = 1)
```

```python
input.head()
```

```python
output.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
model = tree.DecisionTreeClassifier()
```

```python
input_train, input_test, output_train, output_test = train_test_split(input, output, test_size = 0.15)
```

```python
model.fit(input_train, output_train)
```

```python
model.score(input_test, output_test)
```
